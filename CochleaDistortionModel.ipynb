```python
vocalDir = 'C:\\Users\\emahrt\\Documents\\mice_predictions\\stimuli_filter\\distorted_postLPfilter\\'
print vocalDir
vocalFilePattern = vocalDir + '*.call1'
filenames = glob.glob(vocalFilePattern)
filenames = [os.path.splitext(os.path.basename(filename))[0] for filename in filenames]
# print filenames
for vocalName in filenames:
    # vocalName = '23B_21kHz_OLap1ms'

    def PlotDistortion_call1(vocalName, vocalDir, highPassFreq=40000.0, savePath=[]):
        """ plot and save highpass filtered and distorted signal
        :param vocalName: Name of vocalization in a call1 file
        :type vocalName: str
        :param vocalDir: Path to directory containing vocalization file
        :type vocalDir: str
        :param highPassFreq: Highpass frequency
        :type highPassFreq: float
        """
        sd = Signal_Distort.Signal_Distort(vocalDir)  # initialize Signal_Distort class

        # Load vocal signal from .call1 file
        fname = vocalDir + vocalName + '.call1'
        vocalSignal = np.fromfile(fname, dtype='int16')

        #Convert signal to float with range within [-1.0, 1.0]
        nVocalSignal = pcm2float(vocalSignal, np.float32)

        # Highpass filter signal
        nVocalSignal = sd.BandPassFilter(nVocalSignal, lowCut=highPassFreq, set_numtaps=71)

#         # Lowpass filter signal; run twice to remove all high freq energy
#         nVocalSignal = sd.BandPassFilter(nVocalSignal, highCut=lowPassFreq, set_numtaps=71)
#         nVocalSignal = sd.BandPassFilter(nVocalSignal, highCut=lowPassFreq, set_numtaps=71)
        
        # Apply reverberation-distortion filter
        signalReverbDistort = sd.DistortionProducts(nVocalSignal, reverbShift=1., reverbTau=2., echos=5, distortFactor=4.)
        # Scaling to equalize intensity of distorted signal with original.
        signalReverbDistort = signalReverbDistort*(max(abs(nVocalSignal))/max(abs(signalReverbDistort))) 

        # Plot distortion (increase specThreshold to reveal low amplitude noise)
        fig = sd.PlotSpectrogram(signalReverbDistort, figsize=(9,5), title=vocalName, specThreshold=-110)

        # Save figure and call1 file is savePath is given.
        if len(savePath)>0:
            maxAbs_pcm = max(abs(vocalSignal))  # Capture largest integer in the original .call1 file (int16)
            
            pcmScaleDistort = maxAbs_pcm/max(abs(signalReverbDistort))  # Scaling to equalize largest integer (int16) of the distorted signal to match original
            signalReverbDistort_int16 = np.int16(signalReverbDistort*pcmScaleDistort)-1
            signalReverbDistort_int16.tofile(vocalDir + vocalName + '.call1')                   
            
            fig.savefig(vocalDir + vocalName + '.png', bbox_inches='tight')
#             signalReverbDistort_int16.tofile(vocalDir + vocalName + '_distort.call1')
#             fig.savefig(vocalDir + vocalName + '_distort.png', bbox_inches='tight')

    PlotDistortion_call1(vocalName, vocalDir, highPassFreq=40000.0, savePath=vocalDir);
```

**Distortion products for wav files - Rat vocalization**

Source of wav files: http://www.avisoft.com/rats.htm 

Used in: Gao, Patrick P., et al. *BOLD fMRI study of ultrahigh frequency encoding in the inferior colliculus*. NeuroImage (2015).

```python
from scipy.io import wavfile
import thorns.waves as wv

def PlotDistortion_wav(vocalName, dirPath, fs, pre_lowCut=40000.0, pre_highCut=70000.0):
    # load vocal file
    fname = dirPath + vocalName
    fsData,data = wavfile.read(fname)
    maxAbs_pcm = max(abs(data))
    vocalSignal = pcm2float(data, np.float32)
    vocalSignal = sd.BandPassFilter(vocalSignal, lowCut=pre_lowCut, highCut=pre_highCut, set_numtaps=71)
    nVocalSignal = vocalSignal/max(abs(vocalSignal))

    # distort signal
    signalDistort = sd.DistortionProducts(nVocalSignal, reverbShift=1., reverbTau=2., echos=5, distortFactor=1)
    signalDistort = signalDistort*(max(abs(nVocalSignal))/max(abs(signalDistort))) # Scaling to equalize intensity of distorted signal with original.
    
    # Plot Signals and Spectrogram of distorted signal
    fig, (ax1, ax2,ax3) = plt.subplots(3,1, figsize=(18,4))
    fig.suptitle(vocalName, fontsize=14)
    ax1 = plt.subplot(131)
    wv.plot_signal(signalDistort, fs=fs, label='distorted')
    wv.plot_signal(nVocalSignal, fs=fs, label='original')
    plt.legend()
    ax2 = plt.subplot(132)
    Pxx, freqs, bins, im = ax2.specgram(nVocalSignal, NFFT=2**8, Fs=fs, noverlap=10, vmin=-95, cmap = cm.CMRmap)
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')
    plt.ylim((0,80000))

    ax3 = plt.subplot(133)
    Pxx, freqs, bins, im = ax3.specgram(signalDistort, NFFT=2**8, Fs=fs, noverlap=10, vmin=-95, cmap = cm.CMRmap)

    plt.xlabel('Time (s)')
#     plt.ylabel('Frequency (Hz)')
    plt.ylim((0,80000))
    # plt.colorbar(im);
    

dirPath = '/Users/robertpa/Dropbox/Active_Projects/Auditory/cochlea/avisoft_ratVocals/'    # location of data
vocalDir = ''
fs = 214285    # Original sample rate from file description
sd = Signal_Distort.Signal_Distort(vocalDir, sampleRate=fs)

PlotDistortion_wav('rat_60_3.wav', dirPath, fs, pre_lowCut=40000.0, pre_highCut=70000.0)
```
